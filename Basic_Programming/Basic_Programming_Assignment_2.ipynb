1. Write a Python program to convert kilometers to miles?

```python
kms = float(input("Enter the distance in Kilometers: "))

miles = kms/1.60934
print("Converted Distance in Miles:", miles)
```

2. Write a Python program to convert Celsius to Fahrenheit?

```python
celsius = float(input("Enter the temperature in celsius: "))

fahrenheit = 9*celsius/5 + 32
print("Converted Temperature in Fahrenheit:", fahrenheit)
```

3. Write a Python program to display calendar?

```python
import calendar

yy = int(input("Enter the Year: "))
mm = int(input("Enter the month of Year: "))

print(calendar.month(yy, mm))
```

4. Write a Python program to solve quadratic equation?

```python
import math

a = int(input("Enter the varible for quadratic part: "))
b = int(input("Enter the variable for linear part: "))
c = int(input("Enter the constant part: "))

d = (b**2) - 4*a*c

if(d >= 0):
    x1 = (-b + math.sqrt(d))/ (2*a)
    x2 = (-b - math.sqrt(d))/ (2*a)
    print("Solution of Quadratic Equations are", x1, " and", x2)
else:
    print("Complex Solution for this Quadratic Equation.")
```

5. Write a Python program to swap two variables without temp variable?

```python
a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))

a = a + b
b = a - b
a = a - b

print("Value of a: {}".format(a))
print("Value of b: {}".format(b))
```
