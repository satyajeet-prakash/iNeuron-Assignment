1. Write a Python Program to Find the Factorial of a Number?

```python
n = int(input("Enter the number: "))

def factorial(n):
    
    fact = 1
    for i in range(n , 1 , -1):
        fact *= i
    return fact

factorial(n)
```

2. Write a Python Program to Display the multiplication Table?

```python
n = int(input("Enter the Number: "))

def mul_table(n):
    print(n, "* 1 = ", n*1)
    print(n, "* 2 = ", n*2)
    print(n, "* 3 = ", n*3)
    print(n, "* 4 = ", n*4)
    print(n, "* 5 = ", n*5)
    print(n, "* 6 = ", n*6)
    print(n, "* 7 = ", n*7)
    print(n, "* 8 = ", n*8)
    print(n, "* 9 = ", n*9)
    print(n, "* 10 = ", n*10)

mul_table(n)
```

3. Write a Python Program to Print the Fibonacci sequence?

```python
n = int(input("Enter the range: "))

def fibonacci(n):
    a = 0
    b = 1
    print("Fibonacci Number: ", a, b, end=" ")
    
    while(a+b < n):
        print(a + b, end=" ")
        b = a + b
        a = b - a

fibonacci(n)
```

4. Write a Python Program to Check Armstrong Number?

```python
n = int(input("Enter the number: "))

def check_armstrong(n):
    a = n
    sum = 0
    while a >= 1:
        sum += (int(a%10))**3
        a /= 10
    if n == sum:
        return True
    else:
        return False

if check_armstrong(n):
    print("An armstrong number.")
else:
    print("Not an armstrong number.")
```

5. Write a Python Program to Find Armstrong Number in an Interval?

```python
n = int(input("Enter the range: "))
armstrong_list = [i for i in range(2,n) if check_armstrong(i)]
armstrong_list
```

6. Write a Python Program to Find the Sum of Natural Numbers?

```python
n = int(input("Enter the number: "))

def sum_natural_no(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum
sum_natural_no(n)
```
