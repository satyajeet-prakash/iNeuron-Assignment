```python
import numpy as np

arr = np.array([i for i in range(4,30,3)])
arr
```

```python
#1
sum = 0
for i in arr:
    sum += i
print("Summation of Array is", sum)
```

```python
#2
import sys

maxm = -sys.maxsize -1
for i in arr:
    if(maxm < i):
        maxm = i
print("Largest Element in array is", maxm)
```

```python
#3
print(arr[::-1])
```

```python
#4
s_arr = np.array_split(arr, 3)
print("Split Array:",s_arr )

for i in range(1,len(s_arr)):
    temp = s_arr[i-1]
    s_arr[i-1] = s_arr[i]
    s_arr[i] = temp
print("Final Result", s_arr)
```

```python
#5
def check_decresing(arr):
    for i in range(1, len(arr)):
        if(arr[i-1] < arr[i]):
            print("Not a Monotonic Array")
            return
    print("Monotonic Array, Decreasing Array")

def check_increasing(arr):
    for i in range(1, len(arr)):
        if(arr[i-1] > arr[i]):
            print("Not a Monotonic Array")
            return
    print("Monotonic Array, Increasing Array")
    
def check_monotonic(arr, a, b):
    if(b == len(arr)):
        print("Not Monotonic Array, Neither increasing nor decreasing")
    elif(arr[a] > arr[b]):
        check_decreasing(arr)
    elif(arr[a] < arr[b]):
        check_increasing(arr)
    else:
        check_monotonic(arr,a+1, b+1)

check_monotonic(arr, 0, 1)
        
```
